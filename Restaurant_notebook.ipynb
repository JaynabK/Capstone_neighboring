```python

coordinates = pd.DataFrame(columns=['Latitude','Longitude'])
for r in df['Restaurant Name']:
    address = r
    geolocator = Nominatim(user_agent="my-application")
    try:
        location = geolocator.geocode(address)
        latitude = location.latitude
        longitude =location.longitude
    except:
        latitude=0
        longitude=0
    coordinates = coordinates.append({'Latitude':latitude,'Longitude':longitude},ignore_index=True)
df= df.join(coordinates, how ='outer')
df
```

```python
import numpy as np
mask1= df.Latitude <0
mask2= df.Longitude >0
column_name1 = 'Latitude'
df.loc[mask1, column_name1] = 0
column_name2 = 'Longitude'
df.loc[mask2, column_name2] = 0
df= df.replace(0,np.nan)
df= df.replace(df['Latitude']<0,np.nan)
df= df.replace(df['Longitude']>0,np.nan)
# simply drop whole row with NaN in "Borough" column
df.dropna(subset=["Latitude"], axis=0, inplace = True)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)
df.dropna(subset=["Longitude"], axis=0, inplace = True)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)
df
```
